# Tutorial rapido jupyter notebook

## Estrutura

### Aplicativo jupyter notebook

O aplicativo jupyter notebook é uma aplicação cliente-servidor que permite editar e rodar documentos notebook via um browser. Ele também possui um dashboard.

### Notebook Dashboard

O dashboard de notebooks é o componente que é mostrado assim que abrimos o jupyter. Ele é utilizado principalmente para a interação com o sistema de arquivos e controle sobre os documentos notebooks e o aplicativo jupyter notebook.

### Documentos notebook

Os documentos notebook, são arquivos criados utilizando o aplicativo jupyter notebook. Eles podem conter tanto código executavel, dependente de um "notebook kernel", como elementos de texto e gráficos. Podemos dizer que eles são tanto documentos "human-readable" de uma análise quanto um documento executavel que performa a análise

### Notebook kernel

O notebook kernel é uma _engine_ computacional que permite a execução do código contido no documento notebook. Para o curso utilizaremos uma _engine_ python, porém poderiamos utilizar outras se fosse o caso.

## Células

O notebook jupyter é uma IDE baseada em células. Inicialmente temos 4 tipos, que podem ser incrementados com plugins:

### Markdown
    
Células markdown são células de texto como essa. Utilizam formatação do tipo markdown, que é uma extensão do html aceitando inclusive tags html como . Porém o markdown utilizado pelo notebook também foi extendido para aceitar matemática com formatação tipo latex (ver formula abaixo). Finalmente dependendo das das extensões instaladas, podemos inserir resultados de código python no texto.
- Latex -> $$x = y_c^{\frac{2}{7}}$$ 
- Execução python dentro do markdown -> {{2+2}}

### Code
Células contendo código a ser executado na linguagem escolhida no kernel. Para o caso deste curso todos os kernels serão de python, porém poderia ser utilizado para outras linguagens.

### Heading
Deprecado, a forma correta de adicionar cabeçalhos agora é pelo padrao do markdown

### Raw nbconvert
Nbconvert é uma facilidade do jupyter que permite converter o notebook para outros formatos, como por exemplo o html que usamos com as informações de instalação. Essa célula nos permite colocar qualquer coisa que não deve ser convertida na hora dessa transformação. **Não é muito utilizado**
    
Essas células podem ser executadas em qualquer ordem (mesmo células de texto devem ser executadas para que sua formatação apareça), para isso voce pode usar o botao play no menu do topo, que é equivalente ao atalho CTRL+Enter.

```python
def myfunction(vetor):
    vetor_resposta = list()
    for elemento in vetor:
        vetor_resposta.append(elemento+1)
    return vetor_resposta
```

### Recuperando valores anteriores

Dado que o notebook te permite executar as células em qualquer ordem e a re-execução de uma célula, podemos querer saber o que tinha sido executado na ordem. Para isso podemos utilizar as variáveis In e Out

```python
for _index, values in enumerate(In):
    
    print("{}, In: {}".format(_index,values))
    if _index in Out.keys():
        print("Out: {}".format(Out[_index]))
```

### Exibição de gráficos

Usualmente a criação de gráficos em python utiliza matplotlib. Para que esses gráficos apareçam no seu notebook, você precisa utilizar a mágica %matplotlib. Existem dois tipos
- %matplotlib inline: simplesmente joga o gráfico gerado na saida do notebook. 
- %matplotlib notebook: gera um javascript que engloba o gráfico e permite iteração com o gráfico 

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(vetor,vetor)
print("Gráfico inline ")
```

```python
%matplotlib notebook
plt.plot(vetor,vetor,label="x=y")
plt.legend()
plt.show()
print("Gráfico com javascript")
```
